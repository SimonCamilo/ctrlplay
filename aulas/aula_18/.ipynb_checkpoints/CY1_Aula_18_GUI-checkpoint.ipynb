```python
#GUI com grid

#import import_ipynb 
#import importlib  # Por motivos de eficiência, cada módulo é importado apenas uma vez por sessão do interpretador. 
                  # Portanto, se você alterar seus módulos, reinicie o interpretador - ou, se for apenas um módulo 
                  # que deseja testar interativamente, use importlib.reload() para recarregar o módulo:

import CY1_Aula_18_ChatbotBase as pc
from tkinter import *

main_window = Tk()
#importlib.reload(pc)
main_window.title('Dona Maria')
main_window.geometry('1000x700')
main_window.grid()

fontePadrao = ("Arial", "12")
fonte14 = ('Arial','14')

frame = Frame(main_window)
frame.grid()

l_identify = Label(frame, text = 'Insira sua mensagem aqui: ',font=fonte14)
l_identify.grid(row=0, column=0)

e_mensagem = Entry(frame, font=fontePadrao)
e_mensagem.grid(row=0, column=1)

frame2 = Frame(main_window)
frame2.grid(row=1, column=0)
v = StringVar()
Label(frame2, textvariable=v, font=fontePadrao).grid()

nome_maquina = 'Maria'
v.set('Qual é o seu nome? ')
entrada_sugestao = False
entrada_nome_usuario = True
nome_usuario = ''

def roda_Chatbot():
    global entrada_sugestao
    global entrada_nome_usuario
    global historico_conversa
    global nome_usuario
    
    if entrada_nome_usuario:
        nome_usuario = e_mensagem.get()
        saudacao = pc.saudacao_GUI(nome_maquina, nome_usuario)
        historico_conversa = '{}: {}\n'.format(nome_maquina, saudacao)
        v.set(historico_conversa)
        entrada_nome_usuario = False
    
    else:
        texto = e_mensagem.get()
        historico_conversa += '\n{}: {}'.format(nome_usuario, texto)
        v.set(historico_conversa)

        if entrada_sugestao:
            pc.salva_sugestao(texto)
            entrada_sugestao = False
            historico_conversa += '\n Agora aprendi! Vamos continuar nossa conversa... \n'
            v.set(historico_conversa)
        else:
            resposta = pc.buscaResposta_GUI('Cliente: {}\n'.format(texto))
            if resposta == 'Não vem não! Me respeite!':
                historico_conversa += '\n{}: Não vem não! Me respeite!\n'.format(nome_maquina)
                v.set(historico_conversa)
                texto = e_mensagem.get()
                resposta = pc.buscaResposta_GUI('Cliente: {}\n'.format(texto))
            
            elif resposta == 'Ok. Você está fazendo um (a) reclamação, elogio ou sugestão':
                historico_conversa += '\n{}: Ok. Você está fazendo um (a) reclamação, elogio ou sugestão\n'.format(nome_maquina)
                v.set(historico_conversa)
                texto = e_mensagem.get()
                resposta = pc.buscaResposta_GUI('Cliente: {}\n'.format(texto))
                
            else:
                if resposta == 'Me desculpe, não sei o que falar':
                    historico_conversa += '\n Me desculpe, não sei o que falar. O que você esperava? \n'
                    v.set(historico_conversa)
                    entrada_sugestao = True
                else:
                    historico_conversa += '\n' + pc.exibeResposta_GUI(texto, resposta, nome_maquina)
                    v.set(historico_conversa)

Button(frame, text='Clique', command=roda_Chatbot, font=fontePadrao).grid(row=0, column=2)
        

main_window.mainloop()
```

```python
#GUI com Pack

#import import_ipynb 
#import importlib  # Por motivos de eficiência, cada módulo é importado apenas uma vez por sessão do interpretador. 
                  # Portanto, se você alterar seus módulos, reinicie o interpretador - ou, se for apenas um módulo 
                  # que deseja testar interativamente, use importlib.reload() para recarregar o módulo:

import CY1_Aula_18_ChatbotBase as pc
from tkinter import *

class Application:
    
    def __init__(self, master=None): #método construtor
        
        self.img = PhotoImage(file="figura.png")
        self.label_imagem = Label(root,image = self.img, background='red').pack()
        
        self.fontePadrao = ("Arial", "15")
        
        self.primeiroContainer = Frame(master, background='red')
        self.primeiroContainer["pady"] = 10
        self.primeiroContainer.pack()

        self.segundoContainer = Frame(master, background='red')
        self.segundoContainer["padx"] = 20
        self.segundoContainer.pack()

        self.terceiroContainer = Frame(master, background='red')
        self.terceiroContainer["pady"] = 30
        self.terceiroContainer.pack()

        self.titulo = Label(self.primeiroContainer, text='  Chatbot Ctrl+Play  ', background='red', foreground='black')
        self.titulo["font"] = ("Arial", "50", "bold")
        self.titulo.pack()

        self.senhaLabel = Label(self.segundoContainer, text="Insira sua mensagem aqui: ", font=self.fontePadrao, 
                                background='red', foreground='black')
        self.senhaLabel.pack(side=LEFT)

        self.senha = Entry(self.segundoContainer)
        self.senha["width"] = 30
        self.senha["font"] = self.fontePadrao
        self.senha.pack(side=LEFT)
        
        self.autenticar = Button(self.terceiroContainer)
        self.autenticar["text"] = "Clique"
        self.autenticar["font"] = ("Calibri", "14")
        self.autenticar["width"] = 12
        self.autenticar["command"] = self.roda_Chatbot
        self.autenticar.pack()
        
        self.historico_conversa = ''
        self.nome_maquina = 'Maria'
        self.entrada_sugestao = False
        self.entrada_nome_usuario = True
        self.nome_usuario = ''

        self.v = StringVar()
        self.v.set('Qual é o seu nome? ')
        self.vLabel = Label(self.terceiroContainer, textvariable=self.v, font=self.fontePadrao, background='red', foreground='black')
        self.vLabel.pack(side=LEFT)

    def roda_Chatbot(self):
        
        if self.entrada_nome_usuario:
            self.nome_usuario = self.senha.get()
            saudacao = pc.saudacao_GUI(self.nome_maquina, self.nome_usuario)
            self.historico_conversa = '{}: {}\n'.format(self.nome_maquina, saudacao)
            self.v.set(self.historico_conversa)
            self.entrada_nome_usuario = False

        else:
            texto = self.senha.get()
            self.historico_conversa += '\n {}: {}'.format(self.nome_usuario, texto)
            self.v.set(self.historico_conversa)

            if self.entrada_sugestao:
                pc.salva_sugestao(texto)
                self.entrada_sugestao = False
                self.historico_conversa += '\n Agora aprendi! Vamos continuar nossa conversa... \n'
                self.v.set(self.historico_conversa)
            else:
                resposta = pc.buscaResposta_GUI('Cliente: {}\n'.format(texto))
                if resposta == 'Não vem não! Me respeite!':
                    self.historico_conversa += '\n{}: Não vem não! Me respeite!\n'.format(self.nome_maquina)
                    self.v.set(self.historico_conversa)
                    texto = self.senha.get()
                    resposta = pc.buscaResposta_GUI('Cliente: {}\n'.format(texto))
            
                elif resposta == 'Ok. Você está fazendo um (a) reclamação, elogio ou sugestão':
                    self.historico_conversa += '''\n{}: Ok. Você está fazendo um (a) reclamação, elogio ou 
                                                sugestão\n'''.format(self.nome_maquina)
                    self.v.set(self.historico_conversa)
                    texto = self.senha.get()
                    resposta = pc.buscaResposta_GUI('Cliente: {}\n'.format(texto))

                else:
                    if resposta == 'Me desculpe, não sei o que falar':
                        self.historico_conversa += '\n Me desculpe, não sei o que falar. O que você esperava? \n'
                        self.v.set(self.historico_conversa)
                        self.entrada_sugestao = True
                    else:
                        self.historico_conversa += '\n' + pc.exibeResposta_GUI(texto, resposta, self.nome_maquina)
                        self.v.set(self.historico_conversa)

#Programa Principal
root = Tk()
#importlib.reload(pc)
root.title('Chat bot (nao é o ultimo projeto)')
root.geometry('1000x700')
root.configure(background='red')
Application(root)
root.mainloop()
```
