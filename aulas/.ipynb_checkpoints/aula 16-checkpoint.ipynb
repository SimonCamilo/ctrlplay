```python
import requests as r
from bs4 import BeautifulSoup

try:
    url = 'https://www.uol.com.br/start/esport/'
except:
    print('error')
else:
    page = requests.get(url)

soup = BeautifulSoup(page.content, 'html.parser')

lista = ['Free Fire', 'warfare', 'Pokémon']

for paragrafo in soup.find_all('body'):
    for palavra in lista:
        for paragrafo_str in paragrafo.stripped_strings:
            if palavra.upper() in str(paragrafo_str).upper():
                print('NOTICIA SOBRE: {}\n{}\n'.format(palavra.upper(),paragrafo_str))
```

```python
import requests
from bs4 import BeautifulSoup

from tkinter import *
from tkinter import messagebox

class Application:
    def __init__(self, master=None):    
        self.fontePadrao = ('Arial','34')
    
        self.primeiroContainer = Frame(master)
        self.primeiroContainer ['pady'] = 10
        self.primeiroContainer.pack()

        self.segundoContainer = Frame(master)
        self.segundoContainer ['padx'] = 10
        self.segundoContainer.pack()

        self.terceiroContainer = Frame(master)
        self.terceiroContainer ['pady'] = 10
        self.terceiroContainer.pack()
    
        self.titulo = Label(self.primeiroContainer, text= 'Analisador de sites')
        self.titulo['font'] = ('Arial','26','bold')
        self.titulo.pack()

        self.senhaLabel = Label(self.segundoContainer, text = 'site', font = self.fontePadrao)
        self.senhaLabel.pack(side=LEFT)
        self.senha = Entry(self.segundoContainer)
        self.senha['width'] =30
        self.senha['font'] = self.fontePadrao
        self.senha.pack(side=LEFT)

        self.autenticar = Button(self.terceiroContainer)
        self.autenticar ['text'] = 'Verificar'
        self.autenticar ['font'] = ('calibri','14')
        self.autenticar ['width'] = '12'
        self.autenticar['command']=self.verificaSite
        self.autenticar.pack()

        self.mensagem = Label(self.terceiroContainer, text = '', font = self.fontePadrao)
        self.mensagem.pack()

    def verificaSite(self):
        try:
            site = self.senha.get()
            resultado = requests.get(site)
            resposta = resultado.text
            soup = BeautifulSoup(resposta, 'html.parser')
        except:
            messagebox.showinfo(
            title = 'eror',
            message = 'tente denovo')
        else:
            opcao = messagebox.askokcancel(
                title = 'Analisador de sites', 
                message = 'Site: {} \n Pronto para continuar?'.format(site))

            self.mensagem['text'] = soup.title.string
            
            
root = Tk()
Application(root)
root.mainloop()
```

```python
import requests
from bs4 import BeautifulSoup

from tkinter import *
from tkinter import messagebox

class Application:
    def __init__(self, master=None):    
        self.fontePadrao = ('Arial','34')
    
        self.primeiroContainer = Frame(master)
        self.primeiroContainer ['pady'] = 10
        self.primeiroContainer.pack()

        self.segundoContainer = Frame(master)
        self.segundoContainer ['padx'] = 10
        self.segundoContainer.pack()

        self.terceiroContainer = Frame(master)
        self.terceiroContainer ['pady'] = 10
        self.terceiroContainer.pack()
    
        self.titulo = Label(self.primeiroContainer, text= 'Analisador de sites')
        self.titulo['font'] = ('Arial','26','bold')
        self.titulo.pack()

        self.senhaLabel = Label(self.segundoContainer, text = 'assunto da noticia', font = self.fontePadrao)
        self.senhaLabel.pack(side=LEFT)
        
        self.senha = Entry(self.segundoContainer)
        self.senha['width'] =30
        self.senha['font'] = self.fontePadrao
        self.senha.pack(side=LEFT)

        self.autenticar = Button(self.terceiroContainer)
        self.autenticar ['text'] = 'Verificar'
        self.autenticar ['font'] = ('calibri','14')
        self.autenticar ['width'] = '12'
        self.autenticar["command"] = self.verificaSite
        self.autenticar.pack()

        self.historico = ''
        
        self.v = StringVar()
        self.v.set('o que deseja pesquisar? ')
        Label(self.terceiroContainer, textvariable=self.v, font=self.fontePadrao).pack(side=LEFT)

    def verificaSite(self):
        try:
            url = "https://www.cnnbrasil.com.br/"
            resultado = requests.get(url)
            soup = BeautifulSoup(resultado.content, 'html.parser')
            noticia = self.senha.get()
        except:
            messagebox.showinfo(
            title = 'eror',
            message = 'tente denovo')
        else:
            for paragrafo in soup.find_all('body'):
                for paragrafo_str in paragrafo.stripped_strings:
                    if noticia.upper() in str(paragrafo_str).upper():
                        self.historico += 'NOTÍCIA: {}\n'.format(paragrafo_str)
                        self.v.set(self.historico)


root = Tk()
Application(root)
root.mainloop()
```
